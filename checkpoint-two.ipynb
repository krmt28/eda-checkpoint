## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python
#look for correlation
plt.figure(figsize=(15,15))
sns.heatmap(df.corr(), annot=True, cbar=True, cmap="Spectral", fmt=".1f")

plt.savefig('heatmap.png')

#unemployement and covid impact are correlated
#healthcare also made impact in the highest scoring cities
```

```python
#bargrapgh to compare the number of cities that each country has on the list
country_city_counts = df.groupby('Country')['City'].count()
sorted_counts = country_city_counts.sort_values(ascending=False)
plt.figure(figsize=(10, 8))
sorted_counts.plot.bar(color = 'teal')
plt.xlabel('City Count')
plt.ylabel('Country')
plt.savefig('citycountbycountry.png')
plt.title('City Counts by Country')
```
